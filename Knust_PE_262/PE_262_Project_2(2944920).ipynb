### Algorithm

#### (I) Import the necessary libraries
#### (II) Define a function (i.e React_rate)
#### (III) Define the various variables and assign their values
#### (IV) Use Numpy for the range of temperature vaues
#### (V) Define a "list" variable for the reaction rate (i.e k[ ])
#### (VI)  Use the for loop to generate the reaction rate values from its formula( by using the append function to append the individual values)
#### (VII) Plot reaction rate versus absolute temperature (Using subplot)
#### (VIII) Plot rection rate versus the inverse of the absolute temperature
#### (IX) Adjust the sizes of the plots and add some styles to them (i.e color(green), marker etc)
#### (X) Cal the function (i.e React_rate() )

```python
import math
import numpy as np
import matplotlib.pyplot as plt
def React_rate():  # Defining a function with no argument
    fq = 7*(10**16)  # frequency factor
    Ae = 1*(10**5)  # Activation energy in J/mol
    R = 8.314   # Gas constant in J/mol.K
    Ta = np.arange(253,325)  # The absolute temperature in Kelvin(K)
    k = [] # Defining a "list" variable for the reaction rate(1/s)
    for i in Ta:  # Using the for loop on the range of temperatures values
        ka = fq*(math.exp(-Ae/(R*i)))
        k.append(ka)  # appending each values to the rection rate variable
    
    #Plotting reaction rate(1/s) versus absolute temperature(K)
    plt.figure(figsize=[15,10])
    plt.subplot(1,2,1) # Using subplot
    plt.plot(Ta,k, color="green",marker="*")
    plt.xlabel("Absolute temperature($^OK$)")
    plt.ylabel("Reaction rate($s^-1$)")
    plt.title("Reaction rate vs Temperature")
    
    #Plotting the logarithm of reaction rate versus inverse of absolute temperature
    
    plt.subplot(1,2,2)
    plt.tight_layout(pad=10)
    plt.semilogy(1/Ta,k, color="red", marker="*")
    plt.xlabel("$1/^OK$")
    plt.ylabel("Log k($s^-1$)")
    plt.title("Log of reation rate vs inverse of temperature")
React_rate() # Calling the defined function
```
