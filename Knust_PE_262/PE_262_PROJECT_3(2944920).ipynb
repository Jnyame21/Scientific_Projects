## Algorithm

#### Import necessary libraries
#### Define the various variables and assign their values
#### Use np.arange for the diameter values
#### Define variables for storing the Reynold number, Pressure loss and the Friction factor values for water and oil
#### Define three seperate functions for calculating the Reynold number, Pressure loss and Friction factor
#### Plot a graph of pressure loss as a function of pipe diameter for water
#### Plot a graph of pressure loss as a function of pipe diameter for the oil
#### Plot a graph of pressure loss as a function of pipe diameter for both water and oil on the same graph and add a legend
#### Plot two seperate graphs of pressure loss as a function of pipe diameter for both water and oil in the same figure and label it
#### Show the various plots and add necessary attributes

```python
import numpy as np
import matplotlib.pyplot as plt

Avg = 0.001 # Average velocity in meter per seconds
D = np.arange(0.5,10,0.5) # Diameter values in meters(m)

# Q1.

Re = []  # Defining a variable for storing the reynold number values
# COMPUTING THE REYNOLD NUMBER
def Reynold_Number(p,v): # function to calculate the reynold number.It has two arguments,'p' is density and 'v' viscosity
    for i in D: # Using the for loop to iterate through the diameter values and computing for the reynold number
        re = p*Avg*i/v 
        Re.append(re) # Appending the each value into the reynold number variable


PL_water = [] # Defining a variable for storing the pressure loss(in pascal,Pa) values for water
PL_oil = [] # Defining a variable for storing the pressure loss(in pascal,Pa) values for the oil

def Pressure_Loss(): # function to calculate the pressure loss
    # COMPUTING THE PRESSURE LOSS FOR WATER
    Reynold_Number(1000,0.001) # Calling the reynold number function with the two arguments(density and viscosity for water)
    for i,j in zip(Re,D):  # Using the zip function to iterate through the reynold number and diameter values simultaneously 
        if i <= 2300:    # Evaluating the flow regime and calculating the pressure loss for laminar flow  
            f = 64/i  
            pl = f*65*1000*(Avg**2)/(j*2)
        elif i >= 4000: # Evaluating the flow regime and calcutlating the pressure loss for turbulent flow
            f = 0.316*(i**(-1/4))
            pl = f*65*1000*(Avg**2)/(j*2)
        PL_water.append(pl) # Appending each value into the pressure loss variable(thus, PL_water) for water
   
    Re.clear() # Clearing(emptying) the content of the reynold number list 
    
    # COMPUTING THE PRESSURE LOSS FOR THE OIL
    Reynold_Number(870,0.0000107) # Calling the reynold number function with two arguments(density and viscosity for the oil)
    for i,j in zip(Re,D):
        if i <= 2300:
            f = 64/i
            pl = f*65*870*(Avg**2)/(j*2)
        elif i >= 4000:
            f = 0.316*(i**(-1/4))
            pl = f*65*870*(Avg**2)/(j*2)
        PL_oil.append(pl) # Appending each value into the pressure loss variable(thus, PL_oil) for the oil
Pressure_Loss() # Calling the pressure loss function
Re.clear()

f_water = [] # Defining a variable for storing the friction factor values for water
f_oil = [] # Defining a variable for storing the friction factor values for the oil

def Friction_factor(): # Function to calculate the friction factor
    # COMPUTING THE FRICTION FACTOR FOR WATER
    Reynold_Number(1000,0.001)  # Calling the reynold number function with two arguments(density and viscosity for water)
    for i in Re:
        if i <= 2300: # Evaluating the flow regime and calculating the friction factor for laminar flow 
            f = 64/i
        elif i >= 4000: # Evaluating the flow regime and calculating the friction factor for turbulent flow 
            f = 0.316*(i**(-1/4))
        f_water.append(f)  # Appending each value into the friction factor variable(thus, f_water) for water
    
    Re.clear()

    # COMPUTING THE FRICTION FACTOR FOR THE OIL
    Reynold_Number(870,0.0000107) # Calling the reynold number function with two arguments(density and viscosity for the oil)
    for i in Re:
        if i <= 2300:
            f = 64/i
        elif i >= 4000:
            f = 0.316*(i**(-1/4))
        f_oil.append(f) # Appending each value into the friction factor variable(thus, f_oil) for the oil
Friction_factor() # Calling the friction factor function

# Q2.
# Plotting a graph of pressure loss as a function of pipe diameter for water
plt.figure(figsize=[10,5])
plt.plot(D,PL_water, color="b",marker=".")
plt.xlabel("Diameter of pipe(m)")
plt.ylabel("Pressure loss(Pa)")
plt.title("Pressure loss vrs Pipe diameter for water")
plt.grid(True)
plt.show()

# Q3.
# Plotting a graph of pressure loss as a function of pipe diameter for the oil
plt.figure(figsize=[10,5])
plt.plot(D,PL_oil, color="r",marker="*")
plt.xlabel("Diameter of pipe(m)")
plt.ylabel("Pressure loss(Pa)")
plt.title("Pressure loss vrs Pipe diameter for oil")
plt.grid(True)
plt.show()

# Q4.
# Plotting a graph of pressure loss as a function of pipe diameter for both water and oil on the same graph
plt.figure(figsize=[10,5])
plt.plot(D,PL_water, color="b",marker=".")
plt.plot(D,PL_oil, color="r",marker="*")
plt.xlabel("Diameter of pipe(m)")
plt.ylabel("Pressure loss(Pa)")
plt.title("Pressure loss vrs Pipe diameter for both water and oil")
plt.legend(['Water','Oil'], loc="upper right")
plt.grid(True)
plt.show()

# Q5.
# Plotting two seperate graphs of pressure loss as a function of pipe diameter for both water and oil in the same figure
plt.figure(figsize=[10,10])
plt.subplot(2,1,1)
plt.plot(D,PL_water, color="b",marker=".")
plt.xlabel("Diameter of pipe(m)")
plt.ylabel("Pressure loss(Pa)")
plt.title("Pressure loss vrs Pipe diameter for water")
plt.grid(True)
plt.subplot(2,1,2)
plt.tight_layout(pad=5)
plt.plot(D,PL_oil, color="r",marker="*")
plt.xlabel("Diameter of pipe(m)")
plt.ylabel("Pressure loss(Pa)")
plt.title("Pressure loss vrs Pipe diameter for oil")
plt.grid(True)
plt.show()
```
